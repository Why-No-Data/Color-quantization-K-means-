```python
from skimage import io
import numpy as np

centers = np.load('tree.npy')
c = io.imread('tree1.png')
try :
    image = np.zeros((c.shape[0], c.shape[1],3 ), dtype=np.uint8) 
except ValueError:
        print("4")
else:
    
    image = np.zeros((c.shape[0], c.shape[1],4 ), dtype=np.uint8)   
        
for i in range(c.shape[0]):
    for j in range(c.shape[1]):
        image[i,j,:] = centers[c[i,j],:]
        
'''
centers[c[i, j], :]-->>will give us respective cluster of that i--j th position
'''

io.imsave('tree2.png', image)
io.imshow(image)
io.show()
```
